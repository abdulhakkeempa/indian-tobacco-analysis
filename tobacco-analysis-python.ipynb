```python
import pandas as pd

# Specify the path to your Excel file
file_path = "/content/GYTS4.xls"

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

# (Optional) Select specific sheet or columns
# df = pd.read_excel(file_path, sheet_name="Sheet1", usecols=["Col1", "Col2"])

df.head()  # Print the first few rows of the DataFrame
```

```python
import re

# Specify columns of interest
columns_of_interest = [
    "Median age of initiation of Cigarette (in years)",
    "Median age of initiation of Bidi (in years)",
    "Median age of initiation of smokeless tobacco (in years)",
    "Bought Cigarette as individual sticks in past 30 days    (%)",
    "Students who saw anyone using tobacco on mass media in past 30 days  (%)",
]

# Display unique values in each specified column
for col in columns_of_interest:
    print(f"Unique values in {col}:")
    unique_values = df[col].unique()
    print(unique_values)

    # Convert strings to integers (excluding '%' symbol)
    try:
        numeric_values = [int(re.sub(r"\D", "", str(val))) for val in unique_values]
    except ValueError:
        print("No numeric values found in", col)
        continue

    # Replace original values with integers if possible
    if len(numeric_values) > 0:
        df[col] = [int(re.sub(r"\D", "", str(val))) for val in df[col]]
    else:
        print("No numeric values found in", col)
        # Handle empty numeric values as needed (e.g., drop column)

print("Dataframe successfully converted and cleaned!")
```

```python
# Drop unwanted columns
df.drop(columns=["Students who saw anyone using tobacco on mass media in past 30 days  (%)",
                   "Bought Cigarette as individual sticks in past 30 days    (%)"], inplace=True)

# Handle missing values (if necessary)
df.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'State/UT' is the column with state names
states = df['State/UT'].unique()

# Specify the tobacco usage types you're interested in
tobacco_usage_types = ['Current tobacco smokers (%)', 'Current cigarette users (%)', 'Current bidi users (%)', 'Current smokeless tobacco users (%)']

# For each state, get the mean usage for each type
usage_data = {state: [df[df['State/UT'] == state][usage_type].mean() for usage_type in tobacco_usage_types] for state in states}

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# The x locations for the groups
ind = range(len(states))

# Create the bars
for i, usage_type in enumerate(tobacco_usage_types):
    ax.bar(ind, [usage_data[state][i] for state in states], label=usage_type)

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('States')
ax.set_ylabel('Tobacco Usage (%)')
ax.set_title('State-wise Tobacco Usage')
ax.set_xticks(ind)
ax.set_xticklabels(states)
ax.legend()

# Set the y-axis range based on "Ever tobacco users (%)"
min_val = df['Ever tobacco users (%)'].min()
max_val = df['Ever tobacco users (%)'].max()
ax.set_ylim([min_val, max_val])

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

plt.figure(figsize=(30,16))

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'State/UT' is the column with state names
states = df['State/UT'].unique()

# Specify the feature you're interested in
feature = 'Never cigarette smokers susceptible to cigarette use in future (%)'

# For each state, get the mean value for the feature
feature_data = {state: df[df['State/UT'] == state][feature].mean() for state in states}

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# The x locations for the states
ind = range(len(states))

# Create the bars
ax.bar(ind, [feature_data[state] for state in states])

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('States')
ax.set_ylabel(feature)
ax.set_title('State-wise ' + feature)
ax.set_xticks(ind)
ax.set_xticklabels(states)

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

plt.figure(figsize=(30,16))

# Show the plot
plt.show()
```
