# Data Processing

```r
library(readxl)
df <- read_excel("/content/GYTS4.xls")
head(df, 5)
```

```r
# unique values per columns
apply(df, 2, unique)
```

```r
str(df)

# Check for null values
print("Null values in the data frame:")
print(sum(is.na(df)))
```

```r
# Assuming df is your data frame

# Specify the columns of interest
columns_of_interest <- c(
  "Median age of initiation of Cigarette (in years)",
  "Median age of initiation of Bidi (in years)",
  "Median age of initiation of smokeless tobacco (in years)",
  "Bought Cigarette as individual sticks in past 30 days (%)",
  "Students who saw anyone using tobacco on mass media in past 30 days (%)"
)

# Display unique values in each specified column
for (col in columns_of_interest) {
  print(paste("Unique values in", col))
  unique_values <- unique(df[[col]])
  print(unique_values)

  # Convert strings to integers
  numeric_values <- as.integer(gsub("\\D", "", unique_values))

  # Check if the resulting vector is not empty
  if (length(numeric_values) > 0) {
    # Replace original values in the data frame with integers
    df[[col]] <- as.integer(gsub("\\D", "", df[[col]]))
  } else {
    # If the resulting vector is empty, print a message or take appropriate action
    print("No numeric values found.")
  }
}

# Now, df contains the modified values with strings converted to integers
```

```r
df <- na.omit(df)
```

```r
df$"Students who saw anyone using tobacco on mass media in past 30 days  (%)" <- NULL
df$"Bought Cigarette as individual sticks in past 30 days    (%)" <- NULL
```

# Data Visualisation

```r
install.packages("corrplot")
install.packages("choroplethr")
```

```r
library(corrplot)
numeric_df <- df[sapply(df, is.numeric)]
correlation_matrix <- cor(numeric_df)
corrplot(correlation_matrix, method = "circle")
```

```r
df$"Ever tobacco users (%)"
df$"State/UT"
```

```r
# Load the ggplot2 package
library(ggplot2)

# Create the bar plot
ggplot(df, aes(x=`State/UT`, y=`Ever tobacco users (%)`)) +
  geom_bar(stat="identity", fill="steelblue") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  labs(x="State/UT", y="Ever tobacco users (%)", title="Percentage of Ever Tobacco Users by State")
```

```r
library(ggplot2)

ggplot(df, aes(x=`Ever tobacco users (%)`, y=`Current tobacco users (%)`)) +
  geom_point() +
  labs(x="Ever tobacco users (%)", y="Current tobacco users (%)", title="Scatter Plot of Ever vs Current Tobacco Users")
```

```r
# Histogram for 'Ever tobacco users (%)'
ggplot(df, aes(x=`Ever tobacco users (%)`)) +
  geom_histogram(binwidth=5, fill="steelblue") +
  labs(x="Ever tobacco users (%)", title="Histogram of Ever Tobacco Users")

# Histogram for 'Current tobacco users (%)'
ggplot(df, aes(x=`Current tobacco users (%)`)) +
  geom_histogram(binwidth=5, fill="steelblue") +
  labs(x="Current tobacco users (%)", title="Histogram of Current Tobacco Users")
```

```r
library(ggplot2)
library(choroplethr)

# Prepare data
clean_data <- df %>%
  group_by(State/UT) %>%
  summarise(
    ever_users = sum("Ever tobacco users (%)" > 0),
    current_users = sum("Current tobacco users (%)" > 0)
  )

# Create choropleth map
ggplot(clean_data, aes(geometry = State/UT)) +
  geom_choropleth(aes(fill = ever_users, color = current_users)) +
  labs(title = "Ever vs. Current Tobacco Users",
       fill = "Ever Users (%)", color = "Current Users (%)") +
  theme_map()
```
