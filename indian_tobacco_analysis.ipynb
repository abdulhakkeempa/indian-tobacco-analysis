# Data Processing

```r
library(readxl)
df <- read_excel("/content/GYTS4.xls")
head(df, 5)
```

```r
# unique values per columns
apply(df, 2, unique)
```

```r
str(df)

# Check for null values
print("Null values in the data frame:")
print(sum(is.na(df)))
```

```r
# Assuming df is your data frame

# Specify the columns of interest
columns_of_interest <- c(
  "Median age of initiation of Cigarette (in years)",
  "Median age of initiation of Bidi (in years)",
  "Median age of initiation of smokeless tobacco (in years)",
  "Bought Cigarette as individual sticks in past 30 days (%)",
  "Students who saw anyone using tobacco on mass media in past 30 days (%)"
)

# Display unique values in each specified column
for (col in columns_of_interest) {
  print(paste("Unique values in", col))
  unique_values <- unique(df[[col]])
  print(unique_values)

  # Convert strings to integers
  numeric_values <- as.integer(gsub("\\D", "", unique_values))

  # Check if the resulting vector is not empty
  if (length(numeric_values) > 0) {
    # Replace original values in the data frame with integers
    df[[col]] <- as.integer(gsub("\\D", "", df[[col]]))
  } else {
    # If the resulting vector is empty, print a message or take appropriate action
    print("No numeric values found.")
  }
}

# Now, df contains the modified values with strings converted to integers
```

```r
library(ggplot2)
library(choroplethr)

# Prepare data
clean_data <- df %>%
  group_by(State/UT) %>%
  summarise(
    ever_users = sum("Ever tobacco users (%)" > 0),
    current_users = sum("Current tobacco users (%)" > 0)
  )

# Create choropleth map
ggplot(clean_data, aes(geometry = State/UT)) +
  geom_choropleth(aes(fill = ever_users, color = current_users)) +
  labs(title = "Ever vs. Current Tobacco Users",
       fill = "Ever Users (%)", color = "Current Users (%)") +
  theme_map()
```

```r
# Load libraries
library(ggplot2)
library(dplyr)

# Select relevant columns
data_subset <- df %>%
  select(
    "State/UT", "Ever tobacco users (%)", "Current tobacco users (%)",
    "Ever cigarette users (%)", "Current cigarette users (%)",
    "Ever bidi users (%)", "Current bidi users (%)",
    "Ever smokeless tobacco users (%)", "Current smokeless tobacco users (%)"
  )

# Reshape data for easier plotting
data_melted <- data_subset %>%
  melt(id.vars = "State/UT", measure.vars = c(
    "Ever tobacco users (%)", "Current tobacco users (%)",
    "Ever cigarette users (%)", "Current cigarette users (%)",
    "Ever bidi users (%)", "Current bidi users (%)",
    "Ever smokeless tobacco users (%)", "Current smokeless tobacco users (%)"
  )) %>%
  rename(variable = "Tobacco Usage")

# Create the bar plot
ggplot(data_melted, aes(x = State/UT, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  facet_wrap(~ variable) +
  labs(title = "State-wise Tobacco Usage",
       x = "State/UT", y = "Percentage (%)", fill = "Tobacco Usage") +
  theme_bw()
```

```r
# Create empty data frame
reshaped_data <- data.frame("State/UT" = character(), variable = character(), value = numeric())

# Loop through each tobacco usage category
for (col in c(2:9)) {
  current_col <- names(df)[col]
  variable_name <- gsub(" \\(.+\\)$", "", current_col)  # Extract variable name
  reshaped_data <- rbind(reshaped_data,
                            data.frame(
                              "State/UT" = df$`State/UT`,
                              variable = paste(variable_name, "Users"),
                              value = df[, col]
                            )
    )
}

# Add current/ever information
reshaped_data$value <- ifelse(grepl("^Ever ", reshaped_data$variable),
                             reshaped_data$value / 100,
                             reshaped_data$value)
reshaped_data$variable <- gsub("^Ever ", "", reshaped_data$variable)
```
