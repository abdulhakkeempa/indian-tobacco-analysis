```python
df['Students who saw anyone using tobacco on mass media in past 30 days  (%)'] = df['Students who saw anyone using tobacco on mass media in past 30 days  (%)'].replace('63..4', 63.4)
print(df['Students who saw anyone using tobacco on mass media in past 30 days  (%)'].dtype)
```

```python
df['Bought Cigarette as individual sticks in past 30 days    (%)'] = df['Bought Cigarette as individual sticks in past 30 days    (%)'].replace('35,2', 35)
print(df['Bought Cigarette as individual sticks in past 30 days    (%)'].dtype)
```

```python
df['Median age of initiation of Bidi (in years)'] = df['Median age of initiation of Bidi (in years)'].replace('<7.0', 6)
print(df['Median age of initiation of Bidi (in years)'].dtype)
```

```python
df['Median age of initiation of smokeless tobacco (in years)'] = df['Median age of initiation of smokeless tobacco (in years)'].replace('<7', 6)
df['Median age of initiation of smokeless tobacco (in years)'] = df['Median age of initiation of smokeless tobacco (in years)'].replace('--', 0)
df['Median age of initiation of smokeless tobacco (in years)'] = df['Median age of initiation of smokeless tobacco (in years)'].replace('<7.0', 6)
df['Median age of initiation of smokeless tobacco (in years)'] = df['Median age of initiation of smokeless tobacco (in years)'].replace(0, df['Median age of initiation of smokeless tobacco (in years)'].median())
print(df['Median age of initiation of smokeless tobacco (in years)'].dtype)
```

```python
df['Median age of initiation of Cigarette (in years)'] = df['Median age of initiation of Cigarette (in years)'].replace('--', 0)
df['Median age of initiation of Cigarette (in years)'] = df['Median age of initiation of Cigarette (in years)'].replace('<7', 6)
df['Median age of initiation of Cigarette (in years)'] = df['Median age of initiation of Cigarette (in years)'].replace(0, df['Median age of initiation of Cigarette (in years)'].median())
print(df['Median age of initiation of smokeless tobacco (in years)'].dtype)
```

```python
print(f"Old Shape : {df.shape}")
print(df.dropna(inplace=True))
print(f"New Shape : {df.shape}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'df' is your DataFrame and 'State/UT' is the column with state names
states = df['State/UT'].unique()

# Specify the tobacco usage types you're interested in
tobacco_usage_types = ['Current tobacco smokers (%)', 'Current cigarette users (%)', 'Current bidi users (%)', 'Current smokeless tobacco users (%)']

# For each state, get the mean usage for each type and normalize the data
usage_data = {state: [df[(df['Area'] == "Total") & (df['State/UT'] == state)][usage_type].mean() for usage_type in tobacco_usage_types] for state in states}
usage_data = {state: [value/sum(values)*100 for value in values] for state, values in usage_data.items()}

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(15, 10))

# The x locations for the groups
ind = range(len(states))

# Create the bars
bottom = np.zeros(len(states))
for i, usage_type in enumerate(tobacco_usage_types):
    values = [usage_data[state][i] for state in states]
    ax.bar(ind, values, bottom=bottom, label=usage_type)
    bottom += values

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('States')
ax.set_ylabel('Tobacco Usage (%)')
ax.set_title('State-wise Tobacco Usage')
ax.set_xticks(ind)
ax.set_xticklabels(states)
ax.legend()

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

plt.show()
```

## Linear Regression

```python
new_feature_df = df.copy()
```

```python
new_feature_df['Tobacco_Initialisation_Age'] = new_feature_df[['Median age of initiation of Bidi (in years)', 'Median age of initiation of smokeless tobacco (in years)', 'Median age of initiation of Cigarette (in years)']].median(axis=1)
```

```python
for col in new_feature_df.columns:
  if new_feature_df[col].dtype == "object":
    print(f"{col} : {new_feature_df[col].dtype}")
```

```python
X = new_feature_df.drop('Tobacco_Initialisation_Age', axis=1)
y = new_feature_df['Tobacco_Initialisation_Age']
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
```

```python
X = enc.fit_transform(X)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
import numpy as np

np.set_printoptions(suppress=True)

print(np.round(lr.coef_, 5))
print(np.round(lr.intercept_, 5))
```
